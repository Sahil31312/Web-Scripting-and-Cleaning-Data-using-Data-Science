Data Collection Process
</h2>
<p style="font-family:'Arial', sans-serif; font-size:15px; line-height:1.8;">
In this step, I collect employees data from the official University of Pittsburgh at Bradford directory website.
I access the <b>employee directory webpage</b> and retrieve the HTML content of the page.
</p>
<p style="font-family:'Arial', sans-serif; font-size:15px; line-height:1.8;">
Once the HTML is retrieved, I use <b>BeautifulSoup</b> to parse the content and convert it into a structured format that can be easily navigated.
</p>

</div>


```python
url = 'https://www.upb.pitt.edu/directory'
get_data = req.get(url)
get_data = get_data.text
soup = BeautifulSoup(get_data,'html.parser')
soup
```

<!-- Extracting Employee Sections -->
<div style="border-radius:10px; background-color:#d1c4e9; padding:20px; margin-bottom:20px; box-shadow:0 6px 12px rgba(0,0,0,0.2);">
<h3 style="margin-top:0; color:#4527a0; font-family:'Segoe UI', sans-serif;">Extracting Employee Sections</h3>
<p style="font-family:'Arial', sans-serif; font-size:14px; line-height:1.6;">
Here, I extract all the employee entries from the parsed HTML. Each employee's information is stored in a separate HTML block with the class <b>'directory__content'</b>. Using <b>BeautifulSoup</b>, I collect all these blocks into a list called <b>employees_section</b>. Then, I create a new empty BeautifulSoup object and append each employee block into it, creating a consolidated soup that contains only the relevant employee data. This makes it easier to process and clean the data in the next steps.
</p>
</div>


```python
employees_section = soup.find_all(class_='directory__content')
new_soup =BeautifulSoup('','html.parser')
for i in employees_section:
    new_soup.append(i)
new_soup
```

<!-- Extracting Employee Data -->
<div style="border-radius:12px; background: linear-gradient(135deg, #ff9a9e 0%, #fecfef 100%);
            padding:20px; margin-bottom:20px; box-shadow:0 6px 12px rgba(0,0,0,0.2);">
<h3 style="margin-top:0; font-family:'Segoe UI', sans-serif; color:#c2185b;">Extracting Employee Data</h3>
<p style="font-family:'Arial', sans-serif; font-size:14px; line-height:1.6;">
From the collected employee sections, I extract all relevant details including full name, position, role, email, phone, and location. Each employee's information is stored as a dictionary, and all dictionaries are appended to the <b>employees</b> list. This prepares the data for cleaning and conversion into a structured DataFrame.
</p>
</div>


```python
employees = []

for section in employees_section:
    employee = {}

    # Name
    full_name_tag = section.find('p', class_='h3').find('a')
    employee['Full Name'] = full_name_tag.get_text(strip=True)

    # Position
    position_tag = section.find('p', class_='bold-blue')
    employee['Position'] = position_tag.get_text(strip=True)

    # Role
    role_tag = section.find('p', class_='field__item')
    employee['Role'] = role_tag.get_text(strip=True)

    # Email
    address_tag = section.find('address')
    email_tag = address_tag.find('a', href=True) if address_tag else None
    employee['Email'] = email_tag.get_text(strip=True) if email_tag and 'mailto:' in email_tag['href'] else ''

    # Phone
    phone_tag = None
    if address_tag:
        for a_tag in address_tag.find_all('a', href=True):
            if 'tel:' in a_tag['href']:
                phone_tag = a_tag
                break
    employee['Phone'] = phone_tag.get_text(strip=True) if phone_tag else ''

    # Location
    location_tag = None
    if address_tag:
        for p_tag in address_tag.find_all('p'):
            if 'Location:' in p_tag.get_text():
                location_tag = p_tag
                break
    employee['Location'] = location_tag.get_text(strip=True).replace('Location:', '').strip() if location_tag else ''

    employees.append(employee)
```

<!-- Splitting Full Name -->
<div style="border-radius:12px; background-color:#74b9ff; padding:20px; margin-bottom:20px; box-shadow:0 6px 12px rgba(0,0,0,0.2);">
<h3 style="margin-top:0; font-family:'Segoe UI', sans-serif; color:white;">Splitting Full Name</h3>
<p style="font-family:'Arial', sans-serif; font-size:14px; line-height:1.6; color:white;">
The <b>Full Name</b> column is split into <b>Last Name</b> and <b>Name</b> using the comma as a separator. i used <b>str.strip()</b> to remove any extra spaces, resulting in cleaner and more structured data.
</p>
</div>


```python
# Split Full Name by comma
split_p_info = df['Full Name'].str.split(',', expand=True)

# First part → Last Name, Second part → First Name
df['Last Name'] = split_p_info[0].str.strip()
df['Name'] = split_p_info[1].str.strip()
```
